```python
import pandas as pd
df= pd.read_csv("12. Seaborn_Batsman.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.dropna(inplace=True)         
df.info()                       
df.describe()                  
df.isnull().sum() 
```

```python
df.dropna(inplace=True) 
```

```python
sns.histplot(df['RunsScored'], bins=20, color='skyblue')
plt.title('Distribution of Runs Scored')
plt.show()
```

```python
sns.lineplot(x='Year', y='RunsScored', data=df)
plt.title('Runs Scored Over the Years')
plt.show()
```

```python
sns.barplot(x='Opposition', y='RunsScored', data=df, estimator='mean')
plt.title('Average Runs by Opposition')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(x='BattingPosition', y='RunsScored', data=df)
plt.title('Runs by Batting Position')
plt.show()
```
